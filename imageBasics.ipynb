```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = os.path.join("data", "heliconia-8599119.jpg")
image = cv2.imread(path)
```

```python
plt.imshow(image)
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
new_img = image
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
cv2.rectangle(new_img, pt1=(3000, 2000), pt2=(4000, 4000), color=(255,0,0), thickness=50)
plt.imshow(new_img)
```

```python
cv2.imwrite(os.path.join("new_image.jpeg"), new_img)
```
